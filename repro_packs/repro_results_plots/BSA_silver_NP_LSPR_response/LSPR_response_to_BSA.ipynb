## LSPR sensor response to BSA

LSPR biosensors detect a target molecule by monitoring plasmon resonance frequency changes due to the proximity of the target to the metallic nanoparticle. We present calculations that shows how our approach can be used to model the response of LSPR biosensors.

- We use real protein model pdb, pqr, mesh
- Locate 2 BSA at different distance +/- of different axis

Our sensor is a spherical nanoparticle of 8 nm, and the incident electric field is -0.0037 $[C]/([{\unicode{x212B}}^2][\epsilon_0])$
